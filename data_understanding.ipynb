```python
import librosa
import numpy as np
import os
import pandas as pd
```

```python
DATASET_PATH = "/dataset"  # Remember to change your dataset path

# Filter Folder
data_file_list = os.listdir(DATASET_PATH)

print(data_file_list)
```

```python
actor_ids = []
actor_emotions = []
actor_intensitys = []
actor_statements = []
actor_repetitions = []
actor_genders = []
actor_paths = []

for i in data_file_list:
  actor = os.listdir(DATASET_PATH + "/" + i)
  for j in actor:
    id = int((i.split('_')[1]))
    string = j.split('-')

    emotion = int(string[2])
    intensity = int(string[3])
    statement = int(string[4])
    repetition = int(string[5])
    gender = id % 2
    path = DATASET_PATH + "/" + i + "/" + j

    print(f'ID: {id}')
    print(f'Emotion: {emotion}')
    print(f'Intensity: {intensity}')
    print(f'Statement: {statement}')
    print(f'Repetition: {repetition}')
    print(f'Gender: {gender}')
    print(f'Path: {path}')

    actor_ids.append(id)
    actor_emotions.append(emotion)
    actor_intensitys.append(intensity)
    actor_statements.append(statement)
    actor_repetitions.append(repetition)
    actor_genders.append(gender)
    actor_paths.append(path)
```

```python
df = pd.DataFrame({'Actor_ID': actor_ids,
                   'Emotion': actor_emotions,
                   'Intensity': actor_intensitys,
                   'Statement': actor_statements,
                   'Repetition': actor_repetitions,
                   'Gender': actor_genders,
                   'Path': actor_paths})
df
```

```python
df.Emotion.replace({1: 'neutral', 2: 'calm', 3: 'happy', 4: 'sad', 5: 'angry', 6: 'fearful', 7: 'disgust', 8: 'surprised'}, inplace=True)
df.Intensity.replace({1: 'normal', 2: 'strong'}, inplace=True)
df.Statement.replace({1: 'Kids are talking by the door', 2: 'Dogs are sitting by the door'}, inplace=True)
df.Gender.replace({0: 'female', 1: 'male'}, inplace=True)
```
